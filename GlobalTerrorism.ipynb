```python
sns.countplot(x='region_txt',data=Terrorism_df)
plt.xticks(rotation=45)
plt.xlabel('Region of Attacked')
plt.ylabel('No. of Attack')
plt.title('Most Attacked region')
plt.show()
```

```python
# From above visualization we can conclude that North Africa was Attacked more times by terrorist 
```

```python
sns.countplot(x='iyear',data=Terrorism_df)
plt.xticks(rotation=90)
plt.xlabel('Year')
plt.ylabel('No. of Attack')
plt.title('Terrorist Attack per year')
plt.show()
```

```python
# In the year of 2014 most of the terrorist Attacks were happened
```

```python
sns.countplot(x='attacktype1_txt',data=Terrorism_df,palette='PuBuGn_r')
plt.xticks(rotation=45)
plt.xlabel('Attack type')
plt.ylabel('No. of Attack')
plt.title('Type of weapon use Attack')
plt.show()
```

```python
# By above visualization we can conclude that Most of the Attack happen by Explosion(Bombing)
```

```python
sns.countplot(x='targtype1_txt',data=Terrorism_df,palette='Spectral')
plt.xticks(rotation=90)
plt.xlabel('Type')
plt.ylabel('Count')
plt.title('Target of Terrorist')
plt.show()
```

```python
# By Above visualization we can conclude that The private citizens and property are attacked in more number by terrorist 
```
